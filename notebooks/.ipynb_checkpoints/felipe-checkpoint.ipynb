```python
from google.cloud import bigquery
from google.cloud import bigquery_storage

import pandas as pd
pd.set_option('display.max_columns', None)

import matplotlib.pyplot as plt
import seaborn as sns

import sys 
import os 

# Add the parent directory to sys.path 
sys.path.append(os.path.abspath("..")) 
from liquor_app.interface.main import preprocess
```

```python
# Create BigQuery client
# client = bigquery.Client()

# # Create BigQuery Storage client
# bqstorage_client = bigquery_storage.BigQueryReadClient()

# query = """
#         SELECT * 
#         FROM `bigquery-public-data.iowa_liquor_sales.sales`
#         WHERE vendor_number IN (
#             SELECT vendor_number
#             FROM `bigquery-public-data.iowa_liquor_sales.sales`
#             GROUP BY vendor_number
#             ORDER BY SUM(bottles_sold) DESC
#             LIMIT 20
#         );
# """

# Fetch results using BigQuery Storage API
df = preprocess()

# Display unique vendor numbers in the retrieved data
print("Unique vendor numbers in retrieved data:", df["vendor_number"].unique())
print("Number of unique vendors:", df["vendor_number"].nunique())

# Display results
df.head()
```

```python
df.info()
```

```python
# Ensure all values in category_name are strings
df['category_name'] = df['category_name'].astype(str).str.replace(r'\[|\]', '', regex=True).str.strip()

# Group by the category column and sum the bottles sold for each category
category_bottles = df.groupby('category_name')['bottles_sold'].sum().reset_index()

# Optionally, sort the results by bottles sold in descending order for easier viewing
category_bottles = category_bottles.sort_values(by='bottles_sold', ascending=False)
category_bottles
```

```python
# Ensure all values in category_name are strings
df['vendor_name'] = df['vendor_name'].astype(str).str.replace(r'\[|\]', '', regex=True).str.strip()

# Group by vendor_name and sum bottles_sold
vendor_sales = df.groupby('vendor_name')['bottles_sold'].sum().reset_index()

# Sort in descending order
vendor_sales = vendor_sales.sort_values(by='bottles_sold', ascending=False)

vendor_sales
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure date is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Extract the month (YYYY-MM format)
df['month'] = df['date'].dt.to_period('M')

# Group by vendor_number and month, summing bottles_sold
monthly_sales = df.groupby(['vendor_name', 'month'])['bottles_sold'].sum().reset_index()

# Convert period type to string for plotting
monthly_sales['month'] = monthly_sales['month'].astype(str)

# Display the first few rows of the processed data
print(monthly_sales.head())
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x='month', y='bottles_sold', hue='vendor_name', marker='o')

plt.title("Total Sales by Vendor Name (Monthly)")
plt.xlabel("Month")
plt.ylabel("Total Bottles Sold")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title="Vendor Namer", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# df["category"] = df["category"].astype(str).str.replace(r'\.0$', '', regex=True)


# Group by category_name and month, summing bottles_sold
monthly_sales_by_category = df.groupby(['category_name', 'month'])['bottles_sold'].sum().reset_index()

# Convert period type to string for plotting
monthly_sales_by_category['month'] = monthly_sales_by_category['month'].astype(str)

# Display the first few rows of the processed data
print(monthly_sales_by_category.head())
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales_by_category, x='month', y='bottles_sold', hue='category_name', marker='o')

plt.title("Total Sales by Category (Monthly)")
plt.xlabel("Month")
plt.ylabel("Total Bottles Sold")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Group by county_number and month, summing bottles_sold
monthly_sales_by_county = df.groupby(['county', 'month'])['bottles_sold'].sum().reset_index()

# Convert period to string for plotting
monthly_sales_by_county['month'] = monthly_sales_by_county['month'].astype(str)

monthly_sales_by_county.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales_by_county, x='month', y='bottles_sold', hue='county', marker='o')

plt.title("Total Sales by County (Monthly)")
plt.xlabel("Month")
plt.ylabel("Total Bottles Sold")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title="County Number", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Price delta retail_price - state_price

df['price_delta'] = df['state_bottle_retail'] - df['state_bottle_cost']

df.head()
```

```python
df['total_evenue'] = df['price_delta'] * df['bottles_sold']

df.head()
```

```python
# Group by county_number and month, summing total_revenue
monthly_dollar_sales_by_county = df.groupby(['county', 'month'])['total_evenue'].sum().reset_index()

# Convert period to string for plotting
monthly_dollar_sales_by_county['month'] = monthly_sales_by_county['month'].astype(str)

monthly_dollar_sales_by_county.head()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_dollar_sales_by_county, x='month', y='total_evenue', hue='county', marker='o')

plt.title("Total Sales by County $ (Monthly)")
plt.xlabel("Month")
plt.ylabel("Total $ Sold")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title="County Number", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
# Group by category_name and month, summing bottles_sold
monthly_dollar_sales_by_category = df.groupby(['category_name', 'month'])['total_evenue'].sum().reset_index()

# Convert period type to string for plotting
monthly_dollar_sales_by_category['month'] = monthly_sales_by_category['month'].astype(str)

# Display the first few rows of the processed data
print(monthly_sales_by_category.head())
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_dollar_sales_by_category, x='month', y='total_evenue', hue='category_name', marker='o')

plt.title("Total Sales by Category $ (Monthly)")
plt.xlabel("Month")
plt.ylabel("Total $ Sold")
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
